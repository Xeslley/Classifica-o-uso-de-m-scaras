```python
# import the Python Image processing Library
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
import os
import matplotlib as mpl
```

```python
current_path = os.getcwd()
# Create an Image object from an Image
image_path = current_path+"\\only_faces\\"

colorImage  = Image.open(image_path+"mask_weared_incorrect_61.png")


rot30 = colorImage.rotate(30)
rot60 = colorImage.rotate(60)
rot90 = colorImage.rotate(90)
rot120 = colorImage.rotate(120)
rot150 = colorImage.rotate(150)
rot180 = colorImage.rotate(180)
rot210 = colorImage.rotate(210)
rot240 = colorImage.rotate(240)
rot270 = colorImage.rotate(270)
rot300 = colorImage.rotate(300)
rot330 = colorImage.rotate(330)


#image brightness enhancer
enhancer = ImageEnhance.Brightness(colorImage)
#factor = 1 #gives original image, 0.5 #darkens the image, 1.5 #brightens the image
filtered = enhancer.enhance(2.5)#2

#image Contrast enhancer
enhancer = ImageEnhance.Contrast(filtered)
#factor = 1 #gives original image, 0.5 #decrease constrast, 1.5 #increase contrast
filtered = enhancer.enhance(1.4)#1.4

#image Color enhancer
enhancer = ImageEnhance.Color(filtered)
filtered = enhancer.enhance(0.4)

#image Sharpness enhancer
enhancer = ImageEnhance.Sharpness(filtered)
filtered = enhancer.enhance(3)

mpl.rcParams['figure.figsize'] = (20, 20)
#subplot(nrows, ncols, index, **kwargs)
fig = plt.figure()
#fig.tight_layout(pad=3.0)
plt.subplots_adjust(hspace = 1, wspace = 0.2)
plt.subplot(2,7,1)
plt.title('Original image')
plt.imshow(colorImage)

plt.subplot(2,7,2)
plt.title('rotated 30')
plt.imshow(rot30)

plt.subplot(2,7,3)
plt.title('rotated 60')
plt.imshow(rot60)

plt.subplot(2,7,4)
plt.title('rotated 90')
plt.imshow(rot90)

plt.subplot(2,7,5)
plt.title('rotated 120')
plt.imshow(rot120)

plt.subplot(2,7,6)
plt.title('rotated 150')
plt.imshow(rot150)

plt.subplot(2,7,7)
plt.title('rotated 180')
plt.imshow(rot180)

plt.subplot(2,7,8)
plt.title('rotated 210')
plt.imshow(rot210)

plt.subplot(2,7,9)
plt.title('rotated 240')
plt.imshow(rot240)

plt.subplot(2,7,10)
plt.title('rotated 270')
plt.imshow(rot270)

plt.subplot(2,7,11)
plt.title('rotated 300')
plt.imshow(rot300)

plt.subplot(2,7,12)
plt.title('rotated 330')
plt.imshow(rot330)

plt.subplot(2,7,13)
plt.title('filtered')
plt.imshow(filtered)

plt.subplot(2,7,14)
plt.title('Original image')
plt.imshow(colorImage)
```

```python
# https://www.tensorflow.org/api_docs/python/tf/image
```

```python
import tensorflow as tf
```

```python
image_string=tf.io.read_file(image_path+"mask_weared_incorrect_30.png")
image=tf.image.decode_png(image_string,channels=3)
```

```python
def visualize(original, augmented):
  fig = plt.figure()
  plt.subplot(1,2,1)
  plt.title('Original image')
  plt.imshow(original)

  plt.subplot(1,2,2)
  plt.title('Augmented image')
  plt.imshow(augmented)
```

```python
## flip_left_right
```

```python
flipped = tf.image.flip_left_right(image)
visualize(image, flipped)
```

```python
## flip_up_down
```

```python
flipped = tf.image.flip_up_down(image)
visualize(image, flipped)
```

```python
## adjust_saturation
```

```python
saturated = tf.image.adjust_saturation(image, 3)
visualize(image, saturated)
```

```python
## adjust_brightness
```

```python
bright = tf.image.adjust_brightness(image, 0.2)
visualize(image, bright)
```

```python
## adjust_contrast
```

```python
contrast = tf.image.adjust_contrast(image,2)
visualize(image, contrast)
```

```python
saturated_contrast_bright = tf.image.adjust_brightness(image, 0.4)
saturated_contrast_bright = tf.image.adjust_contrast(saturated_contrast_bright,3)
saturated_contrast_bright = tf.image.adjust_saturation(saturated_contrast_bright, 0.4)
visualize(image, saturated_contrast_bright)
```
